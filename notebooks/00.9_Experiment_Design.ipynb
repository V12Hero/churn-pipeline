```python
# :package: Required Libraries
from statsmodels.stats.power import NormalIndPower
import numpy as np
```

```python
# :bar_chart: Parameters
p1 = 0.1 # Proportion in group 1 (control)
p2 = 0.13  # Proportion in group 2 (treatment)
alpha = 0.1  # Significance level
power = 0.9  # Desired power

# :white_check_mark: Calculate effect size for proportions
effect_size = abs(p1 - p2) / np.sqrt((p1 * (1 - p1) + p2 * (1 - p2)) / 2)

# :test_tube: Perform power analysis
analysis = NormalIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1.0, alternative='two-sided')
```

```python
# :pushpin: Output: Required sample size per group
print(f"Required sample size per group: {np.ceil(sample_size):0.0f}")
```

```python
# PASO 1: Librerías necesarias
from statsmodels.stats.power import NormalIndPower
import numpy as np
import pandas as pd
from itertools import product
from pyDOE3 import ff2n  # alternativa para diseños factoriales si se requiere
from sklearn.preprocessing import LabelEncoder
```

```python
# PASO 2: Cálculo del tamaño de muestra por grupo
p1 = 0.1
p2 = 0.12
alpha = 0.05
power = 0.95
effect_size = abs(p1 - p2) / np.sqrt((p1*(1-p1) + p2*(1-p2)) / 2)
analysis = NormalIndPower()
n = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1.0)
print(f"Tamaño de muestra por grupo necesario: {np.ceil(n)}")
```

```python
# PASO 3: Definir las variables (mensajes)
v1 = ["69/169","99/199","No offer"]
v2 = ["Single Image","Carrousel"]
v3 = ["Morning","Afternoon", "Evening"]
v4 = ["Urgecny","Reminder","USP","Educacional","Appreciation","Location","Funny"]
vs = [v1, v2, v3, v4]
levels = [len(v) for v in vs]
```

```python
sum(levels)
```

```python
# PASO 4: Generar todas las combinaciones posibles
combinations = list(product(*vs))
df_combinations = pd.DataFrame(combinations, columns=[f"v{i+1}" for i in range(len(vs))])
print(f"Total de combinaciones posibles: {len(df_combinations)}")

# PASO 5: Seleccionar combinaciones óptimas (aquí de forma simple con muestreo aleatorio)
n_trials = sum(levels) + 1  # cambiar según cantidad que quieras testear
df_selected = df_combinations.sample(n=n_trials, random_state=1).reset_index(drop=True)

# PASO 6: Construcción de mensajes
df_selected["mensaje"] = df_selected.apply(lambda row: ' '.join([str(e) for e in row.values]), axis=1)

# PASO 7: Ver mensajes generados
for i, mensaje in enumerate(df_selected["mensaje"], 1):
    print(f"{i}. {mensaje}")
```
