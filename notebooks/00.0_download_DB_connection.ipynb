```python
def get_all_data_from_table(cursor, tablename):

    cursor.execute(f"""
    SELECT 
    TOP 1 *
    FROM MACDB.dbo.{tablename}
    """)
    rows = cursor.fetchall()
    cols = [col[0] for col in cursor.description]
    if "ModifiedOn" in cols:
        cursor.execute(f"""
            SELECT 
                MIN(ModifiedOn)
            FROM MACDB.dbo.{tablename}
        """)
        rows = cursor.fetchall()

        min_date = rows[0][0]
        max_date = "2025-08-01" # datetime.date.today()

        dates_list = pd.date_range(min_date, max_date, freq='1ME')

        tables_result_list = []
        for date in tqdm.tqdm(dates_list):
            cursor.execute(f"""
                SELECT 
                    *
                FROM MACDB.dbo.{tablename}
                WHERE YEAR(ModifiedOn) = {date.year} and Month(ModifiedOn) = {date.month}
            """)
            rows = cursor.fetchall()

            df = pd.DataFrame.from_records(rows, columns=[col[0] for col in cursor.description])
            if df.shape[0] > 0:
                tables_result_list.append(df)

            result_df = pd.concat(tables_result_list).reset_index(drop=True)
    else:
        cursor.execute(f"""
            SELECT 
                *
            FROM MACDB.dbo.{tablename}
        """)
        rows = cursor.fetchall()

        df = pd.DataFrame.from_records(rows, columns=[col[0] for col in cursor.description])

        result_df = df

    return result_df
```

```python
for key, value in tablename_dict.items():
    print(key, value)
    try:
        df = get_all_data_from_table(cursor, value)
        df.to_parquet(f"{key}.parquet")
        print(df.shape)
    except pyodbc.ProgrammingError as e:
        print(e)
        continue
```

```python
import pyodbc
from pathlib import Path

BASE_OUT = Path.cwd()

def get_all_data_from_invoice_table(cursor, tablename, table_items_name):
    out_inv_dir   = BASE_OUT / tablename
    out_items_dir = BASE_OUT / f"{tablename}Items"
    out_inv_dir.mkdir(parents=True, exist_ok=True)
    out_items_dir.mkdir(parents=True, exist_ok=True)
    print(f"[init] Saving invoice parquet to: {out_inv_dir}")
    print(f"[init] Saving items  parquet to: {out_items_dir}")

    cursor.execute(f"""
        SELECT MIN(ModifiedOn) FROM MACDB.dbo.{tablename}
    """)
    _rows = cursor.fetchall()

    min_date = "2025-08-01"
    max_date = "2025-10-01"

    dates_list = pd.date_range(min_date, max_date, freq='1ME')
    print("Processing dates:", dates_list)

    for date in tqdm.tqdm(dates_list):
        cursor.execute(f"""
            SELECT *
            FROM MACDB.dbo.{tablename}
            WHERE YEAR(ModifiedOn) = {date.year}
              AND MONTH(ModifiedOn) = {date.month}
        """)
        rows = cursor.fetchall()
        invoice_df = pd.DataFrame.from_records(rows, columns=[col[0] for col in cursor.description])

        if not invoice_df.empty:
            inv_path = out_inv_dir / f"{date.year}{date.month:02}.parquet"
            print(f"Saving invoice data → {inv_path} ({invoice_df.shape[0]} rows)")
            invoice_df.to_parquet(inv_path)
        else:
            print(f"No invoice rows for {date.year}-{date.month:02}")

        cursor.execute(f"""
            SELECT a.*
            FROM MACDB.dbo.{table_items_name} a
            INNER JOIN (
                SELECT InvoiceID, ModifiedOn
                FROM MACDB.dbo.{tablename}
            ) b ON a.InvoiceID = b.InvoiceID
            WHERE YEAR(b.ModifiedOn) = {date.year}
              AND MONTH(b.ModifiedOn) = {date.month}
        """)
        rows = cursor.fetchall()
        items_df = pd.DataFrame.from_records(rows, columns=[col[0] for col in cursor.description])

        if not items_df.empty:
            items_path = out_items_dir / f"{date.year}{date.month:02}.parquet"
            print(f"Saving items  data → {items_path} ({items_df.shape[0]} rows)")
            items_df.to_parquet(items_path)
        else:
            print(f"No items rows for {date.year}-{date.month:02}")

    return True
```

```python
tablename_dict = {
    "invoice_PAC": ("v_PAC_Invoice", "v_PAC_InvoiceItems"),
    "invoice_PE": ("v_Invoice", "v_InvoiceItems"),
}
```
