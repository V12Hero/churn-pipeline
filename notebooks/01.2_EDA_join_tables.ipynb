```python
import numpy as np
import pandas as pd
```

```python
branch_df = pd.read_parquet("branch.parquet", engine="fastparquet")
vehicle_df = pd.read_parquet("vehicle.parquet", engine="fastparquet")
customer_df = pd.read_parquet("customer.parquet", engine="fastparquet")
```

```python
branch_df.to_parquet("branches.parquet", engine="fastparquet", index=False)
vehicle_df.to_parquet("vehicles.parquet", engine="fastparquet", index=False)
customer_df.to_parquet("customers.parquet", engine="fastparquet", index=False)
```

```python
display(branch_df.head())
display(vehicle_df.head())
display(customer_df.head())
```

```python
pd.merge(
    customer_df.drop(columns=["CreationOn", "ModifiedOn"]),
    vehicle_df.drop(columns=["CreationOn", "ModifiedOn"]),
    on="CustomerID"
).head()
```

```python
print(vehicle_df.shape)
vehicles_cols = set(vehicle_df.columns) - set(["StationBrand"])
vehicle_df = vehicle_df.drop_duplicates(subset=vehicles_cols, keep="first")
print(vehicle_df.shape)
```

```python
vehicle_df["VIN"] = np.where(vehicle_df["VIN"].astype(str).str.len() != 17, np.nan, vehicle_df["VIN"])
vehicle_df["VIN"] = np.where(
    (vehicle_df["VIN"].isnull()) & (vehicle_df["PlateNumber"].astype(str).str.len() == 17),
    vehicle_df["PlateNumber"],
    vehicle_df["VIN"]
)
vehicle_df["PlateNumber"] = np.where(vehicle_df["PlateNumber"].astype(str).str.len() < 6, np.nan, vehicle_df["PlateNumber"])
vehicle_df["PlateNumber"] = np.where(vehicle_df["PlateNumber"].astype(str).str.len() > 7, np.nan, vehicle_df["PlateNumber"])
```
